```python
#попробую написать визуализацию движения робота из второго задания, у него очень простой язык - продвижение на 0.5 по команде "вперед" и 
#поворот на -90 градусов по команде "поворот"
#ещё наверное надо сделать ему какое-то задание начального положения и направления. положение буду задавать координатами, направление - север, запад, юг, восток - то, 
#куда он поедет по команде вперед
```

```python
class Robo:
    """
    Класс, помогающий запилить движение робота по заданной программе
    """
    def __init__(self,position:tuple = (0,0), direction:str = "север")->None:
        """
        инициализация. задаем начальные коодинаты и направление
        """
        self.__X = position[0]
        self.__Y = position[1]
        self.__possible_directions = {"север":(0,1),"восток":(1,0),"юг":(0,-1),"запад":(-1,0),}
        self.__program = []
        self.__leg_lenght = 0.5
        if direction.lower() in self.__possible_directions:
            self.__direction = direction
        else: #поехали на север если какую-то дичь задали вместо направления
            self.__direction = "север"
        print(f"Робот установлен в ({self.__X}, {self.__Y}) и готов ехать на {self.__direction}")
    
    def set_program(self, program: str)->None:
        """
        прочитать и переварить программу
        """
        lowered_program = program.lower()
        list_program = lowered_program.split()
        if set(list_program) == {"вперед", "поворот"}:
            self.__program = list_program[:]
            print("Программа прочитана успешно")
            #print(self.__program)
        else:
            print("Введенная программа содержит недопустимые инструкции")

    def __rotate(self)->None:
        """
        вспомогательный метод для команды "поворот"
        что-то тут плохо, но получше придумать не могу
        даже один раз придумал, потом опять забыл
        """
        if "север" == self.__direction:
            self.__direction = "восток"
        elif "восток" == self.__direction:
            self.__direction = "юг"
        elif "юг" == self.__direction:
            self.__direction = "запад"
        elif "запад" == self.__direction:
            self.__direction = "север"    
    def __do_step(self)->None:
        """
        вспомогательный метод для команды "вперед"
        """
        self.__X += self.__leg_lenght*(self.__possible_directions[self.__direction])[0]
        self.__Y += self.__leg_lenght*(self.__possible_directions[self.__direction])[1]
    
    def move(self)->tuple :
        """
        робот выполняет одно движение по программе
        """
        if self.__program:
            action = self.__program.pop(0)
        if action == "вперед":
            self.__do_step()
        if action == "поворот":
            self.__rotate()
        return (self.__X,self.__Y)            
    
    
    def is_job_done(self)->bool :
        """
        сигнализирует об окончании выполнении программы
        """
        result = False
        if len(self.__program) == 0:
            result = True
        return result    

#проверка работоспособности
#robo2 = Robo((1,1), "юг")
#robo2.set_program("вперед вперед поворот вперед поворот")
#while not robo2.is_job_done():
#    print(robo2.move())
```
