# Расширте поведение классов. Сначала напишите тест, затем реализуйте метод.

1. Расширьте класс автомобиль, реализовав метод ускорения автомоиля на 5 км в час. И метод остановки автомобиля.

2. Расширьте класс Student, создав метод добавления новой оценки и метод вычисления среднего бала.

3. Создайте класс Shape для представления различных геометрических фигур(круг, прямоугольник треугольник). Реализуйте в нем методы вчисления площали и периметра.

# И еще

5. Изменить ранее написнаны классы геометрических фигур так чтобы они были наслдниками класса Shape.

6. Создайьте класс для учета студентов в учебном заведении. Реализуйте методы добавления нового студента(указав имя, возраст, ...), отчисление студента из учебного завдения и вывод списка всех студентов.

7. Измените класс учебного заведния чтобы они внутри себя для хранния информации о студентах использовал класс Student.

8. Создайте класс Polygon, который представляет многоугольник. У него должен быть атрибут для хранения списка вершин. Добавьте методы для вычисления площади и периметра многоугольника.

Созлдайте класс для работы с матрицами. Который будет представлять матрицы и поддреживать основные операции сложение, вычитание умножение транспонирование. Затем добавьте методы для вычисления определителя матрицы и вычисления обраной матрицы.

Разарботайте классы Warehouse Product Order для управления складскими запасами. Реализуйте функцонал по добавлению и списанию остатков, учету остатокво на складе и формированию заказа.
