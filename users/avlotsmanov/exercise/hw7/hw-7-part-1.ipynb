```python
# дан отрывок кода
a_haiku = [
    "Всё глазел на них",
    "Сакуры цветы, пока",
    "Шею не свело"
]
```

```python
def print_haiku(haiku, func):
    for line in haiku:
        print(func(line))
```

```python
def up(line):
    return line.upper()
    
def srez(line):
    return line[2::3]
    
def red(line):
    return '\033[31m' + line + '\033[0m'
```

```python
#заглавными буквами
print_haiku(a_haiku, up)
print()
#каждый третий символ строк
print_haiku(a_haiku, srez)
print()
#красным цветом
print_haiku(a_haiku, red)
```

Задание 4. Напишите простой пользовательский интерфейс к заданию 3.
Пользователю должно выдаваться сообщение о возможностях вывода (меню), и приглашение на ввод
идентификатора действия или символа `q` для выхода, например:

Возможные действия:
1 - вывод хайку заглавными буквами
2 - вывод только третьих символов каждой строки хайку
3 - вывод хайку красным цветом
Введите номер действия или q для выхода: 

После ввода идентификатора действия, необходимо выполнить соответствующее действие 
и снова вывести меню с запросом на ввод. в случае некорректного идентификатора действия 
вывести соответствующее сообщение и снова вывести меню. После ввода q завершить работу с
простым пользовательским интрефейсом

```python
def print_options():
    while True:
        print("="*30)
        key = input(
        ''' \nВыберите действие, которые хотите сделать.
        Возможные действия:
        1 - Вывод хайку заглавными буквами
        2 - Вывод только третьих символов каждой строки хайку
        3 - Вывод хайку красным цветом
        4 - Вывод каждого слова строки хайку разным цветом
        5 - Вывод второго слова каждой строки хайку жирным шрифтом
        6 - Вывод синим цветом, строчным шрифтом
        q - для выхода
        ''')
        print("="*30)
        if key == '1':
            print_haiku(a_haiku, up)
        elif key == '2':
            print_haiku(a_haiku, third)
        elif key == '3':
            print_haiku(a_haiku, red)
        elif key == '4':
            print_haiku(a_haiku, rainbow_words)
        elif key == '5':
            print_haiku(a_haiku, bold_second)
        elif key == '6':
            print_haiku(a_haiku, blue_lower)
        elif key == 'q':
            print('Выход.')
            break
        else:
            print('Запрос не может быть обработан, попробуйте еще раз')
```

```python
#0 - черный, 1 - красный, 2 - зеленый, 3 - желтый, 4 - синий, 5 - фиолетовый, 6 - бирюзовый, 7 - белый
color = [
    "\033[30m",
    "\033[31m",
    "\033[32m",
    "\033[33m",
    "\033[34m",
    "\033[35m",
    "\033[36m",
    "\033[37m",
]
def third(line):
    return line[2]
    
def color_word (color, word):
    return f'{color}{word}\033[0m'
    
def bold(word):
    return '\033[1m' + word + '\033[0m'
    
def low(line):
    return line.lower() 
    
def words(line):   
    return line.split(' ')
```

```python
def rainbow_words(line):
    s = []
    word = words(line)
    for i in range(len(word)):
        s.append(color_word(color[i % len(color)], word[i]))
    return ' '.join(s)

def bold_second(line):
    s = []
    word = words(line)
    for i in range(len(word)):
        if i == 1:
            s.append(bold(word[i]))
        else:
            s.append(word[i])            
    return ' '.join(s)

def blue_lower(line):
    return low(color_word(color[4], line))
```
