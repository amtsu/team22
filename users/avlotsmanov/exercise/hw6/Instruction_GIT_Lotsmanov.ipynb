# Последовательность наших действий (команды вводятся в командной строке linux)

1. В первый раз клонируем проект **(git clone)**
2. Каждый следующий раз когда хотим обновить проект (скачать из удаленного репозитория) -  переключаемся на ветку **main** и выполняем команду **git pull**
3. Создаем новую ветку, если нужно (**git branch** название ветки) - нюанс, перед этим всегда надо стараться обновлять ветку **main**, от которой вы почкуетесь
4. Проверяем список локальных веток если нужно  (**git branch** без флагов)
5. Переключаемся на созданную или нужную нам ветку (**git checkout** название ветки)6.
6. Работаем, создаем где нужно папки файлы юпитер ноутбуки
7. Командой **git status** проверяем в каком состоянии у нас находятся файлы
8. Добавляем нужные файлы или папки в индекс (делаем отслеживаемыми) **git add**
9. Делаем  коммит отслеживаемых файлов (находящихся в индексе) **git commit -m ‘коммнетарий к коммиту внутри кавычек’**
10. Работаем дальше, регулярно проверяя в какой ветке сейчас мы находимся **git branch** перед созданием нового коммита. Создаем цепочку коммитов (сколько нужно)
11. Проверяем если нужно различие между коммитами командой **git diff**
12. Проверяем если интересно последовательность наших коммитов командой  **git log**
13. ВАЖНО - командой  **git status** проверяем что мы закоммитили все нужные изменения, тк зальется только то, что закоммичено
14. Заливаем на удаленный репо наши изменения **git push --set-upstream origin название ветки**
15. На этом моменте можно продолжать работать в локальном репо и с пункта 6 до 14 повторять сколько угодно раз
16. Создаем на сайте **github pull request** для вливания вашей ветки в основную
17. Ждем когда его удовлетворят и вольют нашу ветку в **main**, после этого на локальном репо можно удалить нашу ветку **git branch -d название ветки**
18. В этот момент на вашей локальной машине не будет тех файлов и изменений что вы писали в своей отдельной ветке, также их не будет в ветке **main**, тк она обновлялась до того как их туда влили.
19. Проверяем какие ветки есть на локальном репо **git branch**
20. Переключаемся на ветку main **git checkout main**
21. Затягиваем все изменения с удаленного репо на локальный с момента нашего прошлого пула **git pull**
22. Ура, все наши труды которые мы делали в отдельной ветке появились в ветке **main**, а также появились труды других ребят, чьи ветки тоже влились в вашу.
23. Теперь работаем по кругу!
