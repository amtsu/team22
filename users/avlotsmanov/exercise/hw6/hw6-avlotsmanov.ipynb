3. Напишите программу для робота, который сходит за хлебом из дома в
булочную. Робот умеет поворачивать направо на 90 градусов командой «Поворот»
и передвигаться на 0.5 м командой «Вперед». Робот имеет язык
программирования с поддержкой переменных, математических операторов,
операторов условий и циклов. Булочник сам положит в него хлеб. Нужно
добраться до булочной и вернуться.


```python
#команда "Вперед"
move = 0.5
#команда "Поворот"
rotate = 90

turn_left = rotate * 3
turn_right = rotate * 1

length_1 = 25
length_2 = 22.5
length_3 = 8

way = 0
direction = 0
There_Back_Again = True
# Путь в булочную
if way == 0:
    direction = direction + turn_left
while way <= length_1:
    way = way + move
if way == length_1:
    direction = direction + turn_right
    way = way + move
while length_1 < way <= length_1:
    way = way + move
if way == length_2:
    direction = direction + turn_right
    way = way + move
while length_2 < way <= length_3:
    way = way + move
if way == length_3 and There_Back_Again = True:
    direction = direction + turn_left
    There_Back_Again = False
#пришли
#идем обратно
if way == length_3 and There_Back_Again = False:
    direction = direction + turn_left
```
