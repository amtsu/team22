```python
class Product:
    """
    Класс для представления продукта.
    """
    def __init__(self, name, price, units, quantity):
        self.name = name
        self.price = price
        self.units = units
        self.quantity = quantity

    def __eq__(self, other):
        return self.name == other.name and self.price == other.price

    def __hash__(self):
        return hash((self.name, self.price))

    def __str__(self):
        return f"{self.name}; {self.price}; {self.units}; {self.quantity}"

class Warehouse:
    """
    Класс для управления складскими запасами.
    """
    def __init__(self):
        self.products = set()

    def add_product(self, new_product):
        for product in self.products:
            if new_product == product:
                product.quantity += new_product.quantity
                break
        else:
            self.products.add(new_product)

    def remove_product(self, product_name, quantity):
        for product in self.products:
            if product.name == product_name:
                if product.quantity < quantity:
                    raise ValueError(f"Недостаточно {product_name} на складе для списания {quantity} единиц.")
                product.quantity -= quantity
                if product.quantity == 0:
                    self.products.remove(product)
                break
        else:
            raise ValueError(f"Продукт с именем {product_name} не найден на складе.")

    def products_names(self):
        return [product.name for product in self.products]

    def products_all_info(self):
        return [str(product) for product in self.products]

class Order:
    """
    Класс для формирования заказа.
    """
    def __init__(self, warehouse):
        self.order_list = set()
        self.warehouse = warehouse

    def add_item(self, product_name, quantity):
        for product in self.warehouse.products:
            if product.name == product_name:
                if product.quantity >= quantity:
                    product.quantity -= quantity
                    order_item = Product(product.name, product.price, product.units, quantity)
                    self.order_list.add(order_item)
                    break
                else:
                    raise ValueError(f"Недостаточно {product_name} на складе для заказа {quantity} единиц.")
        else:
            raise ValueError(f"Продукт с именем {product_name} не найден на складе.")

    def remove_item(self, product_name, quantity):
        for product in self.order_list:
            if product.name == product_name:
                if product.quantity < quantity:
                    raise ValueError(f"Недостаточно {product_name} в заказе для удаления {quantity} единиц.")
                product.quantity -= quantity
                if product.quantity == 0:
                    self.order_list.remove(product)
                for warehouse_product in self.warehouse.products:
                    if warehouse_product.name == product_name:
                        warehouse_product.quantity += quantity
                        break
                break
        else:
            raise ValueError(f"Продукт с именем {product_name} не найден в заказе.")

    def order_all_info(self):
        return [str(product) for product in self.order_list]

# Пример использования
product1 = Product('Apple', 27, 'kg', 300)
product2 = Product('TV', 20000, 'item', 5)
product3 = Product('Water 0,5', 50, 'item', 10000)
product4 = Product('TV', 20000, 'item', 7)

warehouse1 = Warehouse()
warehouse1.add_product(product1)
warehouse1.add_product(product2)
warehouse1.add_product(product3)
warehouse1.add_product(product4)

print("Склад:")
print(warehouse1.products_all_info())

order1 = Order(warehouse1)
order1.add_item('Apple', 2)
order1.add_item('TV', 3)
order1.add_item('Apple', 1)

print("\nЗаказ:")
print(order1.order_all_info())

print("\nСклад после заказа:")
print(warehouse1.products_all_info())

# Удаление товара из заказа
order1.remove_item('Apple', 1)

print("\nЗаказ после удаления товара:")
print(order1.order_all_info())

print("\nСклад после удаления товара из заказа:")
print(warehouse1.products_all_info())
```
