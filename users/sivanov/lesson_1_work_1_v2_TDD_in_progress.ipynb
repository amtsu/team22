```python
# задание 2
import mymoneyrecords
bank = mymoneyrecords.MyMoneyRecords()
moneyrecords = bank.VALUES
desc4moneyrecords= bank.DESC
bank.print_money_lists(desc4moneyrecords,moneyrecords)
#-----------------------------------------
#посчитать сколько потратил за все время
def allTimeSpent():
    allspent = 0
    for amount in moneyrecords:
        if amount < 0:
            allspent = allspent - amount
           # print(allspent)
    return allspent
#-----------------------------------------
# посчитать заработок за все время
def allTimeEarn():
    allearn = 0
    for amount in moneyrecords:
        if amount > 0:
            allearn = allearn + amount
           # print(allspent)
    return allearn
#-----------------------------------------
#вывести траты или зароботок или что-то ещё
def printAllTimeSomething(moneyFunc, funcdesc):
    print('за все время было ' + funcdesc + ': \t ' + str(moneyFunc()))
    return None
#-----------------------------------------
#посчитать самые частые траты
#используем findAllTheSameActions из предыдущего задания
#если будет несколько частых трат, выведет первую 
#ломает этот случай писать.
def findTheMostFrequentExpence():
    uniqueexpence=[]
    allexpence=[]
    freqlist=[]
    mostfrequentexpence=[]
    frequency = 0
    #сформируем список трат
    for i in range(0,len(desc4moneyrecords)):
        if moneyrecords[i]<0:
            allexpence.append(desc4moneyrecords[i])
    #посчитаем частоту вхождений
    findAllOfTheSameActions(allexpence,uniqueexpence, freqlist,0)
    #теперь выберем самое частое действие
    #print(len(uniqueexpence))
    #print(len(freqlist))
    for i in range(0, len(uniqueexpence)):
        if(freqlist[i] == frequency):
            frequency = freqlist[i]
            mostfrequentexpence.append(uniqueexpence[i])
        if(freqlist[i] > frequency):
            frequency = freqlist[i]
            mostfrequentexpence = []
            mostfrequentexpence.append(uniqueexpence[i])
    return ', '.join(mostfrequentexpence)
#--------------------------------------------------------------
#то же самое со словарем
def findTheMostFrequentExpenceDict():
    uniqueexpence=[]
    allexpence={}
    freqlist=[]
    mostfrequentexpence=[]
    frequency = 0
    #сформируем сразу словарь одинаковых трат и посчитаем частоту вхождений
    for i in range(0,len(desc4moneyrecords)):
        if moneyrecords[i]<0:
            if(allexpence.get(desc4moneyrecords[i]) != None):
                count = allexpence.get(desc4moneyrecords[i]) + 1
                allexpence.update({desc4moneyrecords[i]:count})
            else:
                allexpence.update({desc4moneyrecords[i]:1})
    #теперь выберем самое частое действие
    for expencename,expencecount in allexpence.items():
        if(expencecount == frequency):
            frequency = expencecount
            mostfrequentexpence.append(expencename)
        if(expencecount > frequency):
            frequency = expencecount
            mostfrequentexpence = []
            mostfrequentexpence.append(expencename)
    return ', '.join(mostfrequentexpence)
#--------------------------------------------------------------
#то же самое со словарем
def findTheMostFrequentExpenceDict2():
    uniqueexpence=[]
    allexpence={}
    freqlist=[]
    mostfrequentexpence=[]
    frequency = 0
    #сформируем сразу словарь одинаковых трат и посчитаем частоту вхождений
    for i in range(0,len(desc4moneyrecords)):
        if moneyrecords[i]<0:
            if(desc4moneyrecords[i] in allexpence):
                allexpence[desc4moneyrecords[i]] += 1
            else:
                allexpence[desc4moneyrecords[i]] = 1
    #теперь выберем самое частое действие
    for expencename,expencecount in allexpence.items():
        if(expencecount == frequency):
            frequency = expencecount
            mostfrequentexpence.append(expencename)
        if(expencecount > frequency):
            frequency = expencecount
            mostfrequentexpence = []
            mostfrequentexpence.append(expencename)
    return ', '.join(mostfrequentexpence)
#--------------------------------------------------------------

#--------------------------------------------------------------
#fillMoneyListWithSomeShit()
#findTheMostFrequentExpenceDict()
```

# №3 Напишите функцию, которая перебирает числа от 1 до 15 (включительно) и возвращает список из чисел кратных 3.

```python
# перебор чисел от 1 до maxnum не включая, 
#и вывод тех что кратны base
import batchtestcall
def even3(maxnum, base):
    Result = []
    if(base <= 0): # вот это пришлось добавить из-за 4 теста
        return Result
    for i in range(1,maxnum):
        if (i%base==0):
            Result.append(i)
    #print(Result)
    return Result
#-----------------------------------------------
def even3test1():
    resultoftest = even3(0,0)
    if (resultoftest == []):
        return 'done'
    else:
        return 'not done'
    
def even3test2():
    resultoftest = even3(1,0)
    if (resultoftest == []):
        return 'done'
    else:
        return 'not done'
def even3test3():
    resultoftest = even3(1,1)
    if (resultoftest == []):
        return 'done'
    else:
        return 'not done'
def even3test4():
    resultoftest = even3(15,0)
    if (resultoftest == []):
        return 'done'
    else:
        return 'not done'
def even3test5():
    resultoftest = even3(16,3)
    if (resultoftest == [3, 6, 9, 12, 15]):
        return 'done'
    else:
        return 'not done'
def even3test6():
    resultoftest = even3(15,3)
    if (resultoftest == [3, 6, 9, 12]):
        return 'done'
    else:
        return 'not done'
def even3test7():
    resultoftest = even3(-15,3)
    if (resultoftest == []):
        return 'done'
    else:
        return 'not done'
def even3test8():
    resultoftest = even3(8,1)
    if (resultoftest == [1,2,3,4,5,6,7]):
        return 'done'
    else:
        return 'not done'
#----------------------------------------------------------
batchtestcall.batch_test_call('even3test',8)
even3(16,3)
```
