```python
person={'name' : 'Ivan' , 'age' : 23 , 'city' : 'Moscow'}
if person['age'] >= 18:
    print('совершеннолетний')
else:
    print('несовершеннолетний')

set_1={1,2,3,4,5} 
set_2={4,5,6,7,8,9}
if len(set_1) > len(set_2):
    print('первое больше второго')
elif len(set_1) < len(set_2):
    print('первое меньше второго')
else:
    print('они равны')

fruits_quantity={'яблоко': 5, 'банан': 1, 'апельсин': 1}

if 'груша' in fruits_quantity:
    print('груша уже есть')
else:
    fruits_quantity['груша']=3

for name , price in fruits_quantity.items():
    if price > 1.5:
        print(f' {name} дорогой товар')

seven11={'яйца': 40, 'батон хлеба' : 20}

if 'яйца' in seven11 and 'батон хлеба' in seven11:
    seven11['яйца']= seven11['яйца']-10
    seven11['батон хлеба']= seven11['батон хлеба']-1
    print('купил 1 хлеб, осталось', seven11['батон хлеба'])
    print('купил 10 яиц, осталось', seven11['яйца'])
else:
    print('или хлеба или яиц нет')

students={'Иванова': 22, 'Петров': 13, 'Сидоров': 23}
if students['Иванова'] >= 18:
    print('совершеннолетняя')
else:
    print('несовершеннолетняя')
    
    
    
    
```
