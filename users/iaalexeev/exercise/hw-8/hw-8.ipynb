```python
def triangle_type(a: float, b: float, c: float) -> str:
    """
    Функция, определяющая тип треугольника по длинам его сторон
    """
    result = "не треугольник"
    triangle_exist = (a < (b + c)) and (b < (a + c)) and (c < (a + b))
    if triangle_exist:
        sides = [a, b, c]
        sides.sort()
        acuteTriangle = (sides[0]**2 + sides[1]**2) > sides[2]**2
        obtuseTriangle = (sides[0]**2 + sides[1]**2) < sides[2]**2
        rightTriangle = abs((sides[0]**2 + sides[1]**2) - sides[2]**2) <= 0.01

        if acuteTriangle:
            result = "остроугольный"
        elif obtuseTriangle:
            result = "тупоугольный"
        elif rightTriangle:
            result  = "прямоугольный"
    return result

def test_acute():
    """
    тестируем остроугольность треугольника
    """
    input = [12,12,5] # такие стороны у остроугольного треугольника 
    expected =  "остроугольный"
    assert expected == triangle_type(*input)

def test_tup():
    """
    тестируем тупоугольностьь треугольника 
    """
    input = [5,6,9]
    expected = "тупоугольный"
    assert expected == triangle_type(*input)

def test_abc():
    """
    тестируем буквенный треугольника 
    """
    input = ['один','два','три']
    expected = "не треугольник"
    assert expected == triangle_type(*input)

def test_int():
    """
    тестируем числа треугольника 
    """
    input =  [1.93452,0.1,0.000000000000000001]
    expected = "не треугольник"
    assert expected == triangle_type(*input)

def test_zil():
    """
    холодильник
    """
    input =  ['холодильник']
    expected = "не треугольник"
    assert expected == triangle_type(*input)

def py_test():
    total_ok=0
    total_not=0
    try: 
        test_acute()
    except AssertionError:
        print ("тест не пройден")
        total_not+=1
    else:
        print ("тест пройден")
        total_ok+=1

    try: 
        test_int()
    except AssertionError:
        print ("тест не пройден")
        total_not+=1
    else:
        print ("тест пройден")
        total_ok+=1

    try: 
        test_zil()
    except TypeError:
        print ("тест не пройден")
        total_not+=1
    else:
        print ("тест пройден")
        total_ok+=1

    try: 
        test_abc()
    except AssertionError:
        print ("тест не пройден")
        total_not+=1
    else:
        print ("тест пройден")
        total_ok+=1

    try: 
        test_tup()
    except AssertionError:
        print ("тест не пройден")
        total_not+=1
    else:
        print ("тест пройден")
        total_ok+=1

    print(f' Выполнено тестов: {total_ok}')
    print(f' Не выполнено тестов: {total_not}')

py_test()


    
```
