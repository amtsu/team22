Расширте класс для работы с матрицами, загрузкой и сохранением. Придумаейте тесты.

```python
import numpy as np 
import json 
class Build_Matrix:

    def __init__(self,matrix_1=None,matrix_2=None):
        self.matrix_1=np.array(matrix_1)
        self.matrix_2=np.array(matrix_2)
        

    #сложение матриц
    def summa(self,matrix_f,matrix_s):
        matrix_1=np.array(matrix_f)
        matrix_2=np.array(matrix_s)
        if len(matrix_1)!=len(matrix_2):
            raise ValueError('матрицы разной длины')
        else:
            result = np.add(matrix_1, matrix_2)
            return f'Сумма:\n {result}'

    # вычитание матриц 
    def subtract(self,matrix_f,matrix_s):
        matrix_1=np.array(matrix_f)
        matrix_2=np.array(matrix_s)
        if len(matrix_1)!=len(matrix_2):
            raise ValueError('матрицы разной длины')
        result = np.subtract(matrix_1, matrix_2) 
        return f'Вычитание:\n {result}'

    # умножение матрицы на число 
    def multi_on_int(self,matrix_f,integer):
        matrix=np.array(matrix_f)
        if type(integer)!= int:
            raise ValueError ('введите число ,а не то что вы там ввели')
       
        result = matrix * integer
        return f'Умножение на число:\n {result}'
    # умножение двух матриц 
    def multi_matrix(self,matrix_f,matrix_s):
        matrix_1=np.array(matrix_f)
        matrix_2=np.array(matrix_s)
        if len(matrix_1)!=len(matrix_2):
            raise ValueError('матрицы разной длины')
        result = matrix_1 @ matrix_2
        return f'Умножение двух матриц:\n {result}'

    #транспонирование
    def transport(self,matrix_f):
        matrix=np.array(matrix_f)
        result = matrix.T
        return f'Транспонированная матрица: {result}'

    #вычеселние определителя матрицы 
    def reverse_info(self,matrix_f):
        matrix=np.array(matrix_f)
        determinant = np.linalg.det(matrix)
        if determinant == 0:
            return "У этой матрицы нет обратной"
        else:
            return f'у этой матрицы есть обратная. Определитель матрицы: {determinant:.1f}'

    #вычисление обратной матрицы 
    def reverse(self,matrix_f):
        matrix=np.array(matrix_f)
        result = np.linalg.inv(matrix) 
        return f'Обратная матрица: {result}'
    
    def add_matrix(self,name,matrix_1,matrix_2=None):
    # добавление матрицы в файл
        if matrix_2==None:
            name_file=name+'.json'
            with open (name_file, 'w') as f_w:
                add_matrix={'первая матрица': matrix_1}
                json.dump(add_matrix,f_w)
        else:
            name_file=name+'.json'
            with open (name_file, 'w') as f_w:
                add_matrix={'первая матрица': matrix_1 , 'вторая матрица': matrix_2}
                json.dump(add_matrix,f_w)   

class Matrix(Build_Matrix):

    def __init__(self,file_name):
        self.file=file_name

    @staticmethod
    def load(n):
        ''' загрузка матрицы из файла json'''
        fn=n+'.json'
        with open(fn ,'r') as read:
            matrix=json.load(read)
            one=matrix['первая матрица']
            two=matrix['вторая матрица']
            return [one,two]
        
    def sum (self):
        ''' сумма '''
        rf=self.file+'result_summa.txt'
        with open (rf ,'w') as f:
            f.write(self.summa(self.load(self.file)[0],self.load(self.file)[1]))

    def sub(self):
        '''вычитание'''
        rf=self.file+'_result_subtract.txt'
        with open (rf ,'w') as f:
            f.write(self.subtract(self.load(self.file)[0],self.load(self.file)[1]))
    
    def multi(self,integer=None):
        if integer!=None:
            rf=self.file+'_result_multi_int.txt'
            result=[]
            result.append(self.multi_on_int(self.load(self.file)[0],integer))
            result.append(self.multi_on_int(self.load(self.file)[1],integer))
            with open (rf ,'w') as f:
                f.write(str(result))
        else:
            rf=self.file+'_result_multi.txt'
            with open (rf ,'w') as f:
                f.write(self.multi_matrix(self.load(self.file)[0],self.load(self.file)[1]))
            
    def trans(self):
        rf=self.file+'_result_transport.txt'
        result=[]
        result.append(self.transport(self.load(self.file)[0]))
        result.append(self.transport(self.load(self.file)[1]))
        with open (rf , 'w') as f:
            f.write(str(result))

    def det(self):
        rf=self.file+'_result_det.txt'
        result=[]
        result.append(self.reverse_info(self.load(self.file)[0]))
        result.append(self.reverse_info(self.load(self.file)[1]))
        with open (rf , 'w') as f:
            f.write(str(result))

    def rev(self):
        rf=self.file+'_result_reverse.txt'
        result=[]
        result.append(self.reverse(self.load(self.file)[0]))
        result.append(self.reverse(self.load(self.file)[1]))
        with open (rf , 'w') as f:
            f.write(str(result))
```

```python
import random
import json


class Warehouse:

    '''общий класс с работой с продукатми на складе'''

    TOTAL=0

    def __init__(self, ware_house, basket):
        self.ware_house = ware_house
        self.basket = basket
        

    def load(self):
        file = self.ware_house + '.json'
        with open(file, 'r') as read:
            array = json.load(read)
        return array

    def add_product(self, product, count, price):

        '''добавляет продукт на склад'''

        file = self.ware_house + '.json'
        new_product = {product: {'количество': count, 'цена': price}}
        array = self.load()
        array.update(new_product)
        with open(file, 'w') as add:
            json.dump(array, add)

    def warehouse_info(self):

        '''показывает обстановку на складе'''

        array = self.load()
        print([a for a in array.items()])

    def delete_product(self, product):

        '''удаление товара со склада'''

        file = self.ware_house + '.json'
        array = self.load()
        if product in array:
            del array[product]
            with open(file, 'w') as delete:
                json.dump(array, delete)
        else:
            raise ValueError('такого продукта нет на складе')

    def counter(self, product, count):

        '''меняет количество товаров на складе'''
        if product not in self.load():
            raise ValueError('такого продукта нет')
        array = self.load()
        new_count = array[product]['количество']
        file = self.ware_house + '.json'

        if count == 0:
            raise ValueError('число должно быть больше нуля')

        elif new_count + count < 0:
            array[product]['количество'] = 0

        else:
            new_count = array[product]['количество']
            array[product]['количество'] = new_count+count

        with open(file, 'w') as ct:
            json.dump(array, ct)

    def pricer(self, product, price):

        '''меняет цену товаров на складе'''
        if product not in self.load():
            raise ValueError('такого продукта нет')
        array = self.load()
        new_price = array[product]['цена']
        file = self.ware_house + '.json'

        if price == 0 or new_price + price < 0:
            array[product]['цена'] = 'бесплатно'

        else:
            new_price = array[product]['цена']
            array[product]['цена'] = new_price + price

        with open(file, 'w') as pt:
            json.dump(array, pt)

    @staticmethod
    def add_id():

        '''создает номер заказа и файл'''

        a=1
        b=100
        num = random.randint(a, b)
        ii='order '+ str(num) + '.txt'
        with open (ii,'w'):
            print(f'ORDER ID: {num}')
        
    @classmethod
    def total_price(cls,file,product,count):

        '''считает общую цену'''

        with open(file, 'r') as read:
            array = json.load(read)

        price = array[product]['цена'] * count
        cls.TOTAL+=price
        return cls.TOTAL
    
    @classmethod
    def total(cls,order_id):

        '''костыль для подсчета цены '''
        
        o_id = 'order ' + str(order_id) + '.txt'
        
        with open (o_id, 'r') as read:
            add_total=read.read()

        add_total+=f'TOTAL: {str(cls.TOTAL)}'

        with open (o_id , 'w') as write:
            write.write(add_total)

    def order_product(self, order_id, product, count):

        '''составляет заказ и кладет продукты в корзину (убирая купленное со склада)'''

        array = self.load()
        new_count=-count
        self.counter(product,new_count)
        file_name = self.ware_house + '.json'

        if count <= 0:
            raise ValueError('количество продуктов должно быть больше нуля')
        elif product not in array:
            raise ValueError('такого продукта нет на складе')
        elif array[product]['количество']==0 or array[product]['количество']-new_count<0:
            raise ValueError('товар кончился, или попробуйте взять меньше товара. Товара осталось: ',array[product]['количество'])
        
        o_id = 'order ' + str(order_id) + '.txt'
        price = array[product]['цена'] * count
        basket = (f'Product: {product} \nCount: {count} \nPrice: {price}\n \n')

    
        with open (o_id, 'r') as read:
            all=read.read()

        all+=basket
        
        with open (o_id , 'w') as write:
            write.write(all)
            self.total_price(file_name,product,count)
```
