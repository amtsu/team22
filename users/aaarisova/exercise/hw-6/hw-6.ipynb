#### 2. Напишите программу для робота, который сходит за хлебом из дома в булочную. Робот умеет поворачивать направо на 90 градусов командой «Поворот» и передвигаться на 0.5 м командой «Вперед». Больше ничего не умеет. Булочник сам положит в него хлеб. Нужно добраться до булочной и вернуться.

```python
def turn_left():
    for i in range(3):
        print('Поворот направо')
    
def forward(distance):
    for i in range(distance*2):
        print('Вперед')

def turn_right():
    print('Поворот направо')


def turn_around():
    for i in range(2):
        print('Разворот')
    

turn_left()
forward(25) 
turn_right() 
forward(23)
turn_right()
forward(8)
# обратно домой:
turn_around()
forward(8)
turn_left()
forward(23)
turn_left()
forward(25)
```
