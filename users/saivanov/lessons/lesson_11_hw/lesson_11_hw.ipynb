```python
#---------------------------------------------------------------------------------
# эта штука тоже пригодится
def sprint(some_list, delimiter=u'\n'):
    Out = ''
    for item in some_list:
        if(type(item) == type([])):
            sprint(item,delimiter)
        else:    
            Out += unicode(item) + delimiter
    print(Out.rstrip(delimiter)) 
    return None
```

```python
#-----------------------------------------------------------
#найти действия, законченные в тот же день
def findImmidiatelyDoneActions():
    selectedactions = []
    count = len(todo)
    for i in range(0,count):
        if timetobedone[i] == timedone[i]:
            selectedactions.append(str(i) + u': ' + todo[i] + ' ' + timedone[i])
    for anaction in selectedactions:
        print(anaction)
    return None
#-------------------------------------------
#найти все одинаковые действия
def findAllOfTheSameActions(actionlist, uniqactions, uniqcount, allowprint=1):
    for action in actionlist:
        i=0
        while i < len(uniqactions):
            if action == uniqactions[i]:
                uniqcount[i] = uniqcount[i] + 1
                break
            i=i+1
        if i == len(uniqactions):
            uniqactions.append(action)
            uniqcount.append(1)  
    if allowprint ==1:
        for i in range(0, len(uniqactions)):
             print(u'действие "' + uniqactions[i]+ u'" было выполнено ' + str(uniqcount[i]) + u' раз')
    return None
                    
```
