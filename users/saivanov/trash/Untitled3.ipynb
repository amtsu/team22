```python
def triangleCheckType(a: float, b: float, c: float) -> None:
    if (a + b) > c:
        sides = [a, b, c]
        sides.sort()
        acuteTriangle = (sides[0]**2 + sides[1]**2) > sides[2]**2
        obtuseTriangle = (sides[0]**2 + sides[1]**2) < sides[2]**2
        rightTriangle = abs((sides[0]**2 + sides[1]**2) - sides[2]**2) <= 0.01

        if acuteTriangle:
            print(f"Thе triangle with sides {a, b, c}  is acute")
        elif obtuseTriangle:
            print(f"Thе triangle with sides {a, b, c} is obtuse")
        elif rightTriangle:
            print(f"Thе triangle with sides {a, b, c} is right")
            
    else:
        print(f"The triangle with sides {a, b, c} doesn't exist")

triangleCheckType(3, 4, 5)
triangleCheckType(1000, 2, 3)
triangleCheckType(9, 5, 6)
```

```python
def calc_sum_of_seq_slice(n1, n2):
    k = max(n1, n2)
    n = min(n1, n2)
    return (k*(k+1) - n*(n-1))/2

print(calc_sum_of_seq_slice(-4,-9))
```

```python
def getSummFromRange(n1: int, n2: int) -> int:
    if n1 > n2:
        start = n2
        stop = n1
    else:
        start = n1
        stop = n2
    
    return sum(list(range(start, stop + 1)))

n1 = -4
n2 = -9
print(f"The sum of the numbers between {n1} and {n2} is {getSummFromRange(n1, n2)}")
```

```python
def comparison(first_set: set, second_set) -> str:
    result = ""
    if first_set == second_set:
        result = "они равны"
    elif first_set > second_set:
        result = "первое множество больше"
    else:
        result = "второе множество больше"
    return result 

comparison({1,2,3}, {4,5,6})
```

```python
def is_it_palindrom(string1):
    string2 = []
    for char in string1.lower():
        if str(char).isalnum() == True:
            string2.append(char)
        else: 
            continue
    if string2[::-1] == string2:
        print(f'Cтрока "{string1}"- \033[93mпалиндром\033[0m')
    else:
        print(f'Cтрока "{string1}" - \033[93mне палиндром\033[0m')
    return

string1 = '1asDfg hhgfds,a1'
string2 = 'asdfghhgfdfdfdsa'
is_it_palindrom(string1)
is_it_palindrom(string2)
```
