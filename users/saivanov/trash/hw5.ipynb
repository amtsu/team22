Задание 2.<br><hr>
Задайте переменную a_poem типа строка, со следующим содержимым:
<code>
За стеклом лежал Питон,
Большой и толстый, как батон.
Вверх пополз,
Затем вернулся,
Круглым бубликом свернулся.
</code>
- Создайте переменную a_tuple типа кортеж, заполнив ее содержимым строки a_poem так, чтобы каждый символ строки a_poem стал отдельным элементом a_tuple
- Создайте переменную another_tuple типа кортеж, заполнив ее содержимым строки a_poem так, чтобы каждое слово строки a_poem стал отдельным элементом another_tuple. Знаки препинания прилипнут к словам, не страшно.
- Создайте переменную one_more_tuple типа кортеж, каждым элементом которой является отдельная строка строфы из a_poem.
- Выведите на экран элементы и их количество для кортежей a_tuple, another_tuple и one_more_tuple.

```python
#решение можно писать тут
a_poem = """За стеклом лежал Питон,
Большой и толстый, как батон.
Вверх пополз,
Затем вернулся,
Круглым бубликом свернулся."""
a_tuple = tuple(a_poem)
print(a_tuple, len(a_tuple))
another_tuple = tuple(a_poem.split())
print(another_tuple, len(another_tuple))
one_more_tuple = tuple(a_poem.split('\n'))
print(one_more_tuple, len(one_more_tuple))
```
