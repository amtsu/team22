#### Задача
Создайте функцию, которая будет проверять введенную пользователем строку на наличие повторяющихся символов.
   - Если такие символы найдены, программа должна выводить их перечень и количество повторов для каждого.

#### Модель
Вход: строка . Выход: Словарь, ключи – буквы, значения – количество букв в строке. 

### Алгоритм
1.	Ввод строки
2.	Создаем пустой словарь с выходными данными
3.	Преобразуем строку в список
4.	Выполняем цикл по каждому символу строки
5.	Если символ не содержится в словаре result – добавляем в result символ и значение присваиваем 1.
6.	Если символ содержится в в словаре result – создаем переменную num, равную значению в словаре result. Присваиваем значению текущей буквы значение num + .


### Код

def mult_string(string1):
    result = {}
    s = list(string1)
       
    for i in range(len(s)):
        if s[i] not in result:
            result[s[i]] = 1
        else:
            num = result[s[i]] + 1
            result[s[i]]= num
    
    return result
