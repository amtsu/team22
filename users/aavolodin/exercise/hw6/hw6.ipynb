1. В JupiterLab запустить Terminal, перейти в папку githud/teem22/.
2. Загрузить из репозитория ветки main (предварительно перейдя в нее "git checkout main") последние изменения на свой ПК "git pull".
3. В ветке main создать командой "git branch <имя>" ветку для выполнения ДЗ и перейти в неё с помощью команды "git checkout <имя>.
4. Перейти в папку /users/<Имя>/exercise и создать в ней папку для выполнения ДЗ командой "mkdir".
5. Создать файл для выполнения ДЗ командой "vi <название файла> или открыть Notebook.
6. Приступить к выполнению ДЗ.
7. Командой "git status" проверить статус репозитория после выполнения ДЗ.
8. Сохранить ДЗ командой "git add <имя файла>.
9. Командой "git status" проверить, сохранение нужного нам файла.
10. Создать коммит командой  "git commit -m '<Описание изменений>'".
11. Отправить ветку на мердж коандой "git push".
