## Напишите инструкцию для участника наших курсов по работе с git при выполнении одной домашней работы.

1) зайти в терминал
2) переключиться на main: git checkout main
3) подтянуть изменения: git pull
4) создать ветку для ДЗ: git checkout -b rg-yusupov-hw6
5) перейти в папку cd ~/user/exercises/
6) создать и перейти в папку для ДЗ: mkdir hw6 && cd hw6 
7) создаем файлы: touch hw-6.ipynb
8) открываем в редакторе: vi hw-6.ipynb
9) после выполнения задания делаем коммит git add . && git commit -am "описание изменений"
10) если ДЗ не закончено возвращаемся к пункту 8, если закончено идем дальше
11) делаем git push
12) идем на гит хаб делаем пул реквест
13) если ПР не принят возвращаемся к пунку 8, если принят идем дальше
14) радуемся выполненному ДЗ
