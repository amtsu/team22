# Пример ка кработать с словарями

```python
slovar = {
    'home': 10,
    'car': 'volvo',
    'citis': ['Moscow', 'kazan', 'spb'],
    'notebook': {
        'memory': '100g',
        'cpu': 2,
    }
}

print(slovar)

print(slovar['notebook'])

slovar['temp'] = 'work'

print(slovar)

del slovar['citis']
print(slovar)

#keys
#values
#items

print(slovar.keys())

print(slovar.values())

print(slovar.items())

print('---')
for k, l in slovar.items():
    print('key =', k, ' value =', l)
```
