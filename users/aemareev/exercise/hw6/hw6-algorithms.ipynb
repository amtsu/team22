## 1. Разберите предложенным на занятии методом (составление модели — алгоритм — программа — анализ результата) одно из заданий предыдущих домашних работ, вызвавшее сложности при решении.

### Задача
Создайте функцию для определения типа треугольника по длинам сторон. Функция принимает длины сторон, возвращает "остроугольный, прямоугольный, тупоугольный или не треугольник".
### Модель
Вход:
Три числа, длины сторон треугольника (a, b, c).
Выход:
Строка, тип треугольника.
Связи:
1) Сумма длин двух любых сторон треугольника всегда больше длины третьей стороны, иначе это не треугольник.
2) Если квадрат самой большой стороны больше суммы квадратов двух других его сторон - треугольник тупоугольный.
4) Если квадрат самой большой стороны равен сумме квадратов двух других его сторон - треугольник прямоугольный.
5) Если квадрат самой большой стороны меньше суммы квадратов двух других его сторон- треугольник остроугольный.
### Алгоритм
1) Вводим a, b, c.
2) Проверяем, является ли фигура треугольником, для всех 3-х возможных комбинаций сторон по формуле side_1 + side_2 <= side_3.
3) Если хоть одно неравенство выполняется, выводим результат: строку "не треугольник".
4) Если все три неравенства выполняются, переходим к проверке на тип треугольника.
5) Определим, какая сторона самая большая. Для этого создадим список из трех сторон и отсортируем его по убыванию. Первый элемент списка и будет самой длинной стороной, второй и третий элемент - остальные стороны.
6) Проверим, является ли треугольник тупоугольным, подставив элементы списка в формулу elem_1^2 > elem_2^2 + elem_3^2.
7) Если неравенство выполняется, выводим результат: строку "треугольник тупоугольный".
8) Если неравенство не выполняется, делаем проверку на "прямоугольность" по формуле elem_1^2 = elem_2^2 + elem_3^2.
9) Если неравенство выполняется, выводим результат: строку "треугольник прямоугольный".
10) Если неравентсво не выполняется, выводим результат: строку "треугольник остроугольный".
### Программа

```python
def get_triangle_type(a, b, c) -> str:
    if a + b <= c or a + c <= b or b + c <= a:
        return 'Не треугольник!'

    lst = [a, b, c]
    lst.sort(reverse=True)

    if lst[0] ** 2 > lst[1] ** 2 + lst[2] ** 2:
        return "Тупоугольный"
    elif lst[0] ** 2 == lst[1] ** 2 + lst[2] ** 2:
        return "Прямоугольный"
    else:
        return "Остроугольный"


print(get_triangle_type(3, 3, 9))
print(get_triangle_type(4, 3, 6))
print(get_triangle_type(5, 4, 3))
print(get_triangle_type(4, 4, 4))
```
