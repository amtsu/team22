```python
omar = '''(2, "тот большего добьётся," )
(5, "Кто умирал, тот знает, что живёт.")
(1, "Кто битым жизнью был," )
(3, "Пуд соли съевший выше ценит мёд.")
(4, "Кто слёзы лил, тот искренней смеётся,")'''
omar_true = omar .split('\n') #разделила строку по пробелам
omar_true .sort()             #отсортировала по возрастанию
print('\n' .join(omar_true))  #объединила эл-ты списка в одну строку 
```
