### Создайте функцию, которая определяет, является ли введенная дата корректной (дата вводится в виде трех чисел, день, месяц, год. Корректной датой считается существующая комбинация этих трех чисел).

# Входные данные: число, день, месяц.
# Выходные данные: корректность или не корректность даты
# Алгоритм
1. День не может быть меньше 1 или больше 31
2. Месяц не может быть меньше 1 или больше 12
3. Год не может быть меньше 1582 ( по григорианскому календарю) и не больше 2024
4. Исключаем месяцы в которых 30 дней в году: апрель, июнь, сентябрь, ноябрь.
5. Проверяем на високостность года. Где каждый год, который делится на 4 без остатка, является високосным годом, за исключением тех годов, которые делятся на 100 без остатка. Однако, если год делится на 400 без остатка, то он все равно считается високосным.
6. Создаем программу


```python
def check_date(day, month, year):
    if day < 1 or day > 31:
        return False
        
    if month < 1 or month > 12:
        return False
        
    if year < 1582 and year > 2024:
        return False
        
    if month in [4, 6, 9, 11] and day > 30:
        return False
        
    elif  month == 2:
        
        if (year % 4 == 0 and year % 100 != 0 ) or year % 400 == 0:
            
            if day > 29 or day > 28:
                return False
    
    return True   

day = 30        
month = 4
year = 2022
if check_date(day, month, year):
    print("Дата корректна")
else:
    print("Дата некорректна")
```
