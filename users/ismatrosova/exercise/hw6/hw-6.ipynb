```python
#был код
def leap_year_or_not (val):
    if val%4==0:
        print (val,"високосный")
    else:
        print (val, "обычный")

leap_year_or_not(2032)
leap_year_or_not(2031)
leap_year_or_not(2030)
leap_year_or_not(2029)
leap_year_or_not(2028)
```

```python
#стал код
'''
Если год не делится на 4, то он обычный.
Если год делится на 4, но не делится на 100, то он високосный.
Если год делится на 100, но не делится на 400, то он обычный.
Если год делится и на 100, и на 400, то он високосный.
'''
def leap_year_or_not_new (year):
    if year%4==0:
        if year%100==0:
            if year%400==0:
                result = (year, "високосный")
            else:
                result = (year, "обычный")
        else:
            result = (year, "високосный")
    else:
        result = (year, "обычный")
    return(result)

print(f" Какой год? : {leap_year_or_not_new(1999)}")
print(f" Какой год? : {leap_year_or_not_new(1700)}")
print(f" Какой год? : {leap_year_or_not_new(2024)}")
print(f" Какой год? : {leap_year_or_not_new(1812)}")
print(f" Какой год? : {leap_year_or_not_new(1980)}")
print(f" Какой год? : {leap_year_or_not_new(2200)}")
print(f" Какой год? : {leap_year_or_not_new(1996)}")
print(f" Какой год? : {leap_year_or_not_new(0)}")
print(f" Какой год? : {leap_year_or_not_new(4)}")
```

3. Напишите программу для робота, который сходит за хлебом из дома в
булочную. Робот умеет поворачивать направо на 90 градусов командой «Поворот»
и передвигаться на 0.5 ми командой «Вперед». Робот имеет язык
программирования с поддержкой переменных, математических операторов,
операторов условий и циклов. Булочник сам положит в него хлеб. Нужно
добраться до булочной и вернуться.

```python
line_home_corner = 35 -10
line_corner_corner = 22.5
line_corner_bakery = 8
step = 0.5

def turn_90():
    print("Поворот")

def turn_180 ():
    turn_90()
    turn_90()

def turn_270 ():
    turn_180()
    turn_90()

def forward (distanse):
    i_line = 0
    i = 1
    while i_line < distanse:
        print("Вперед", i)
        i_line+=step
        i+=1



turn_90()
forward(line_home_corner)
turn_90()
forward(line_corner_corner)
turn_90()
forward(line_corner_bakery)
turn_180()
forward(line_corner_bakery)
turn_270()
forward(line_corner_corner)
turn_270()
forward(line_home_corner)
turn_270()
```

4. Напишите программу для робота, который сходит за хлебом из дома в
булочную. Робот умеет поворачивать на заданное число градусов, двигаться
вперед со скоростью 0.2м/с, имеет встроенный счетчик времени. Булочник сам
положит в него хлеб. Нужно добраться до булочной и вернуться.

```python
line_home_corner = 35 -10
line_corner_corner = 22.5
line_corner_bakery = 8
velocity = 0.2

def turn(angle):
    print("Поворот на", angle,"градусов")

def forward (distanse):
    all_time = distanse/velocity
    print("Вперед на",all_time, "секунд")



turn(90)
forward(line_home_corner)
turn(90)
forward(line_corner_corner)
turn(90)
forward(line_corner_bakery)
turn(180)
forward(line_corner_bakery)
turn(270)
forward(line_corner_corner)
turn(270)
forward(line_home_corner)
turn(270)
```
