```python
a_haiku = ["Всё глазел на них",
"Сакуры цветы, пока",
"Шею не свело"]

def print_haiku(haiku,func):
    for line in haiku:
        print(func(line))

def f0(line):
    return line

def f1(line):
    return line.upper()

def f2(line):
    return line[::3]

def f3(line):
    return "\033[31m{}\033[0m".format(line)

def f4(line):
    return "\033[34m{}\033[0m".format(line.lower())

def main():
    instuction ='''\nВас приветствует простой пользовательский интерфейс.
Как вы хотите вывести хайку?
Возможные действия:
1 - вывод хайку заглавными буквами
2 - вывод только третьих символов каждой строки хайку
3 - вывод хайку красным цветом
4 - маленькими буквами синим цветом
Введите номер действия или q для выхода: \n'''
    print (instuction)
    preobrazovanie = (input())
    if (preobrazovanie=="0"):
        print_haiku(a_haiku,f0)
    elif(preobrazovanie=="1"):
        print_haiku(a_haiku,f1)
    elif(preobrazovanie=="2"):
        print_haiku(a_haiku,f2)
    elif(preobrazovanie=="3"):
        print_haiku(a_haiku,f3)
    elif(preobrazovanie=="4"):
        print_haiku(a_haiku,f4)
    elif(preobrazovanie=="q"):
        print("до свидания!")
        return;
    else:
        print("Нет такого пробразования")
    main()

main()
```
