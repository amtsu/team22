### Задание №1

<b>Задача из реального мира</b><br>
6. Создайте функцию для определения типа треугольника по длинам сторон (принимает длины сторон, возвращает "остроугольный, прямоугольный, тупоугольный или не треугольник").<br>
<br>
<b>Определение модели</b><br>
Вход: длины сторон треугольника: a, b, c <br>
Выход: тип треугольника:  прямоугольный, остроугольный<br>
Связь: сравнивая сумму наименьших сторон с наибольшей можно определить тип треугольника<br>
<br>
<b>Алгоритм</b>
1. Получить длины сторон треугольника
2. Проверить возможность существования треугольника с такими сторонами
3. Если треугольник возможен:
    1. Отсортировать стороны по возрастанию
    2. Определить тип треугольника: Взять сумму квадратов и сравнить ее с квадратом большей стороны
    3. Вывести на экран тип треугольника
4. Если треугольник невозможен:
    1. Вывести на экран сообщение, что треугольник с такими сторонами не существует

<b>Программа<b>

````python
def triangleCheckType(a: float, b: float, c: float) -> None:
    triangle_exist = (a < (b + c)) and (b < (a + c)) and (c < (a + b))
    if triangle_exist:
        sides = [a, b, c]
        sides.sort()
        acuteTriangle = (sides[0]**2 + sides[1]**2) > sides[2]**2
        obtuseTriangle = (sides[0]**2 + sides[1]**2) < sides[2]**2
        rightTriangle = abs((sides[0]**2 + sides[1]**2) - sides[2]**2) <= 0.01

        if acuteTriangle:
            print(f"Thе triangle with sides {a, b, c}  is acute")
        elif obtuseTriangle:
            print(f"Thе triangle with sides {a, b, c} is obtuse")
        elif rightTriangle:
            print(f"Thе triangle with sides {a, b, c} is right")
            
    else:
        print(f"The triangle with sides {a, b, c} doesn't exist")
````

<b>Результат<b>

````python
triangleCheckType(3, 4, 5)
triangleCheckType(1, 2, 3)
triangleCheckType(9, 5, 6)
triangleCheckType(1000, 1, 1)
````
