4. Вывести название и среднюю цену всех товаров, у которых со временем цена изменилась.

```python
SEREGA_ZAEBAL('''with P
                as (select distinct title, 
                            first_value(price_sale) over (partition by title order by datetime_create) as fp,
                            first_value(price_sale) over (partition by title order by datetime_create desc) as lp 
                    from products_history)
                select P.title, fp, lp, avg(ph.price_sale)
                from P
                join products_history ph on P.title = ph.title
                group by ph.title
                having fp != lp
                ''')
```
