```python
==========================
Задание 1
==========================
Разберите предложенным на занятии методом (составление модели — алгоритм — программа — анализ результата) одно из заданий предыдущих домашних работ, вызвавшее сложности при решении.

Модель
Входные данные: Год (целое число)
Выходные данные: Ответ на вопрос "является ли год високосным" (ДА или НЕТ)
Связи: Определение. Високосный год - год, который делится на 4 без остатка, 
но при этом не делится на 100 без остатка 
(за исключением годов, которые делятся на 400 без остатка).

Алгоритм
Передать год в функцию в качестве аргумента.
Проверить, делится ли год на 4 без остатка и при этом не делится на 100 без 
остатка, или делится на 400 без остатка.
Если условие выполняется, то год високосный.
Вывести результат.
```

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print(f'Год {year} високосный')
    else:
        print(f'Год {year} НЕ високосный')
```

```python
year1 = 1993
year2 = 1056
year3 = 2500
year4 = 2000
```

```python
is_leap_year(year1)
# Год 1993 НЕ високосный
is_leap_year(year2)
# Год 1056 високосный
is_leap_year(year3)
# Год 2500 НЕ високосный
is_leap_year(year4)
# Год 2000 високосный
```
