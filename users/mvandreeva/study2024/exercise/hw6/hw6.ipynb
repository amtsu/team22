```python
# 1. разберите одним из предложенных на занятии способов (составление модели - алгоритм - программа - анализ результата) одно из заданий предыдущих ДР
# 1. Модель: 
# Входные параметры модели: Строка
# Выходные параметры: True/False/None
# Тело: функция проверки строки на палиндромность.
# Строка будет палиндромом, если буквенные символы в ней читаются одинаково слева направо и справа налево.
# Регистр букв и знаки препинани не должны учитываться.
# 2. Алгоритм:
# 1) Проверяем, что переданное в функцию значение -строка.
# 2) Приводим символы строки к одинаковому регистру
# 3) Убираем символы, оставляем только буквы
# 4) Перебираем символы в строке в обратном порядке (реверс)
# 5) Сравниваем реверсированную строку со строкой в прямом порядке (из п.3)
# 6) если строки равны, то исходная строка - палидром (True), если нет - не палиндром (False)
# 3. Программа:
def is_palindrome(some_str):
    """Функция проверки, является ли строка палиндромом"""
    result = False
    if isinstance(some_str, str):
        lower_str = some_str.lower()
        str_filter = filter(lambda x: x.isalpha(), lower_str)
        str_list = list(str_filter)
        reversed_str_list = list(reversed(str_list[:]))
        if str_list != [] and str_list == reversed_str_list:
            print("Палиндром")
            result = True
        else:
            print("Не палиндром")
    else:
        print("Неверный аргумент")
        result = None
    return result

# 4. Анализ результата:
is_palindrome("Ама") #Палиндром
is_palindrome("gfufhbhbjk") #Не палиндром
is_palindrome("А роза упала на лапу Азора") #Палиндром
# Вывод: программа выдала верные результаты
```

```python
from math import ceil
#Путь до булочной в расстрояниях до поворотов (модель программы):
a = 0.5
b = 35 - 10 + 0.5 #25.5
c = 22.5 + 0.5 + 0.5 #23.5
d = 8 + 0.5 #8.5
e = 0.5
step = 0.5
# step_a = a / step # 1
# step_b  = b / step # 51
# step_c = c / step # 47
# step_d = d / step # 17
# step_e = e / step # 1
# cmd_step = "Вперед"
# cmd_turn = "Поворот"

# Программа
distance_to_corner = [a, b, c, d, e]
def robot_go(distance, step):
    cmd_step = "Вперед"
    total_steps = ceil(distance / step)
    direct_way = cmd_step * total_steps
    print(total_steps)
    return direct_way

def robot_turn_rigth():
    cmd_turn = "Поворот"
    # print(cmd_turn)
    return cmd_turn

def robot_turn_left():
    turn_left = robot_turn_rigth() * 3
    # print(turn_left)
    return turn_left

def robot_turn_around():
    turn_around = robot_turn_rigth() * 2
    return turn_around

def go_to_bakery(distances, step):
    way_to_bakery = []
    way_to_bakery.append("Start adventure".upper())
    for i in range(len(distances)-1):
        # robot_go(distance, step)
        way_to_bakery.append(robot_go(distances[i], step))
        way_to_bakery.append(robot_turn_rigth())
    way_to_bakery.append(robot_go(distances[len(distances) - 1], step))
    way_to_bakery.append("Robot at bakery".upper())
    return way_to_bakery

def go_back_home(distances, step):
    way_home = []
    way_home.append(robot_turn_around())
    distances_back = distances[::-1]
    for i in range(len(distances)):
        way_home.append(robot_go(distances[i], step))
        way_home.append(robot_turn_left())
    way_home.append("Home, sweet home".upper())
    return way_home

total_way = go_to_bakery(distance_to_corner, step) + go_back_home(distance_to_corner, step)
print(total_way)
```

```python
# Расстояния до поворотов
a = 0.5
b = 35 - 10 + 0.5 #25.5
c = 22.5 + 0.5 + 0.5 #23.5
d = 8 + 0.5 #8.5
e = 0.5
speed = 0.2
# step_a = a / step # 1
# step_b  = b / step # 51
# step_c = c / step # 47
# step_d = d / step # 17
# step_e = e / step # 1
# cmd_step = "Вперед"
# cmd_turn = "Поворот"

# Программа
distance_to_corner = [a, b, c, d, e]
def robo_turn(angle):
    # turn = "Turn %i" 
    turn = "Turn " + str(angle)
    return turn
    
def go2bakery(distance_list, speed, angle):
    way_to_bakery = []
    way_to_bakery.append("Start adventure".upper())
    time_count = 0
    for distance in distance_list:
        while distance > 0:
            if distance >= speed:
                distance -= speed
                time_count += 1
            else:
                time_count += distance / speed
                distance = 0
        # way_to_bakery.append(f"Move {time_count} seconds")
        # way_to_bakery.append("Move %i seconds" %time_count)
        if time_count >= 60:
            way_to_bakery.append("Move " + str(round((time_count/60), 2)) + " minutes")
        else:
            way_to_bakery.append("Move " + str(round(time_count, 2)) + " seconds")
        way_to_bakery.append(robo_turn(angle))
    way_to_bakery.append("Robot at bakery".upper())
    return way_to_bakery

def go_home(distance_list, speed):
    way_home = []
    way_home.append(robo_turn(180))
    time_count = 0
    distances_back = distance_list[::-1]
    for distance in distances_back:
        while distance > 0:
            if distance >= speed:
                distance -= speed
                time_count += 1
            else:
                time_count += distance / speed
                distance = 0
        # way_home.append(f"Move {time_count} seconds")
        # way_home.append("Move %i seconds" %time_count)
        if time_count >= 60:
            way_home.append("Move " + str(round((time_count/60), 2)) + " minutes")
        else:
            way_home.append("Move " + str(round(time_count, 2)) + " seconds")
        way_home.append(robo_turn(270))
    way_home.append("Home, sweet home".upper())
    return way_home
# turn = f"Turn {angle}"
# turn = "Turn %i" %angle
# print(turn)
total_way = go2bakery(distance_to_corner, speed, 90) + go_home(distance_to_corner, speed)
print(total_way)
```
